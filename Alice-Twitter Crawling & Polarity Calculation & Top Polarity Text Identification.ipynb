```python
import pandas as pd
import numpy as np

# Open the crawled twitter file with sentiment scores, turn it into a data structure to add column and manipulate data 
polarity_df= pd.read_csv('test_en_sentiment.tsv', delimiter= "\t", header=None)
polarity_df.columns= ["TweetID", "Date", "Time", "Language", "NULL", "Polarity", "Subjectivity"]
polarity_df
```

```python
#extracting only the date, Tweet Id, and polarity score
new_polarity_df= polarity_df[["Date", "TweetID", "Polarity"]]
new_polarity_df
```

```python
#filtering out polarity that are "nan"
filtered_df = new_polarity_df[new_polarity_df["Polarity"].notnull()]
filtered_df
```

```python
#identifying the top highest sentiment text
total_high = ""
for i in top_high_sent["TweetID"]:
    x = api.get_status(i).text
    total_high = total_high + x
print(total_high)
```

```python
#remove the hyperlinks
import re
clean_total_high = re.sub(r"http\S+", "", total_high)
clean_total_high
```

```python
#lower text, remove punctuations 
lowered_total_high = clean_total_high.lower()

punctuations = "!#$%&()*,.:;<=>?@[\]^_`{|}~"
no_punct=''
for char in lowered_total_high:
    if char not in punctuations:
        no_punct = no_punct + char
print(no_punct)
lowered_new_total_high= no_punct
```

```python
#split individual words
words_list= lowered_new_total_high.split()
print(words_list)
```
