```python
import pandas as pd
import numpy as np

dataset= pd.read_csv('WORK january_2020.txt', delimiter= "\t", header=None)
dataset.columns= ["Tweet ID", "Date", "Time", "Language", "NULL"]
```

```python
instances= dataset['Date'].value_counts().sort_index()

# The dataset only includes dates on which there were English-language tweets about COVID.  As a result, this
# Series does not contain every day in January 2020.  For dates which are absent, the number of tweets is 0.

print(instances)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import timeit

# Overall tweet distribution

overall_data= sns.displot(dataset.Date, height=5, aspect=6)
overall_data.set_axis_labels("Date (yy/mm/dd)", "Number of Tweets", fontsize=20)
overall_data.set_xticklabels(rotation=30, size=15)
overall_data.set_yticklabels(size=15)
```
