# Tweet Distribution

## Calculation of Tweet Frequency Per Day

```python
import pandas as pd
import numpy as np

# Opens the file and turns it into a data structure which is easier to manipulate

dataset= pd.read_csv('WORK january_2020.txt', delimiter= "\t", header=None)
dataset.columns= ["Tweet ID", "Date", "Time", "Language", "Country"]
```

```python
# Counts the number of times a given value (the date) occurs in the data structure
# and sorts the results by date

instances= dataset['Date'].value_counts().sort_index()
print(instances)
```

```python
# Because the data does not include dates where there were no tweets,
# the following code appends those dates onto 'instances' in order 
# to provide a more accurate x axis later in the data visualization

no_tweets= pd.Series([0, 0, 0, 0, 0, 0, 0, 0], index=["2020-01-01", "2020-01-02", "2020-01-03", "2020-01-05", "2020-01-07", "2020-01-14", "2020-01-15", "2020-01-16"])
full_instances= instances.append(no_tweets)
series= full_instances.sort_index()
series.index.name= 'Date'
series.reset_index()
print(series)
```

```python
# Creating the dataframe which the data visualization will be based upon

full_data= pd.DataFrame([['2020-01-01', 0], ['2020-01-02', 0], ['2020-01-03', 0],
                            ['2020-01-04', 1], ['2020-01-05', 0], ['2020-01-06', 2],
                            ['2020-01-07', 0], ['2020-01-08', 4], ['2020-01-09', 11],
                            ['2020-01-10', 4], ['2020-01-11', 9], ['2020-01-12', 3],
                            ['2020-01-13', 3], ['2020-01-14', 0], ['2020-01-15', 0],
                            ['2020-01-16', 0], ['2020-01-17', 19], ['2020-01-18', 9],
                            ['2020-01-19', 3], ['2020-01-20', 70], ['2020-01-21', 95],
                            ['2020-01-22', 385], ['2020-01-23', 597], ['2020-01-24', 906],
                            ['2020-01-25', 993], ['2020-01-26', 1198], ['2020-01-27', 48438],
                            ['2020-01-28', 147302], ['2020-01-29', 151240], ['2020-01-30', 181564],
                            ['2020-01-31', 186307]],
                          columns=["Date", "Number of Tweets"])

display(full_data)
```

## Visualizing the Tweet Frequency

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualizing the overall data as a line plot

plt.figure(figsize=(20,10))
monthly_distribution= sns.lineplot(x="Date", y="Number of Tweets", data=full_data)
plt.xticks(rotation=30)
plt.xlabel("Date (yy/mm/dd)", size=15)
plt.ylabel("Number of Tweets", size=15)
plt.title("January 2020 Tweet Distribution", size=20)
plt.show()
```

```python
# Visualizing the first 19 days of distribution as a line plot.
# Because the scale of the y axis for the overall data is so large,
# it becomes difficult to visualize smaller fluctations.  The date
# range for this plot was chosen based on the range of the y axis
# (the number of tweets) and the ability to visualize fluctuations
# which would be invisible within a larger range.

plt.figure(figsize=(20,10))
section_1= sns.lineplot(x="Date", y="Number of Tweets", data=full_data[:19])
plt.xlabel("Date (yy/mm/dd)", size=15)
plt.ylabel("Number of Tweets", size=15)
plt.title("January 1, 2020 - January 19, 2020 Tweet Distribution", size=20)
plt.xticks(rotation=30)
plt.show()
```

```python
# Visualizing the data from January 20-26 as a line plot.
# Because the scale of the y axis for the overall data is so large,
# it becomes difficult to visualize smaller fluctations.  The date
# range for this plot was chosen based on the range of the y axis
# (the number of tweets) and the ability to visualize fluctuations
# which would be invisible within a larger range.

plt.figure(figsize=(20,10))
section_2= sns.lineplot(x="Date", y="Number of Tweets", data=full_data[19:26])
plt.xlabel("Date (yy/mm/dd)", size=15)
plt.ylabel("Number of Tweets", size=15)
plt.title("January 20, 2020 - January 26, 2020 Tweet Distribution", size=20)
plt.show()
```

```python
# Visualizing the data from January 27-31 as a line plot.
# The date range for this plot was chosen based on the range of 
# the y axis (the number of tweets) and the ability to visualize 
# fluctuations.

plt.figure(figsize=(20,10))
section_3= sns.lineplot(x="Date", y="Number of Tweets", data=full_data[26:30])
plt.xlabel("Date (yy/mm/dd)", size=15)
plt.ylabel("Number of Tweets", size=15)
plt.title("January 27, 2020 - January 31, 2020 Tweet Distribution", size=20)
plt.show()
```

# Visualization of Individual Sentiment Polarity

```python
# Turns the result of the sentiment analysis provided by Alice
# into a DataFrame with columns for date and sentiment polarity.

sentiment_data= pd.read_csv('WORK Final_individual_polarity_dataset.csv', delimiter= ",", header=None)
sentiment_data.columns= ["Tweet No", "Date", "Polarity"]

# Cleans up the DataFrame in preparation for visualization

sentiment_data= sentiment_data.drop([0])
sentiment_data["Polarity"]= pd.to_numeric(sentiment_data["Polarity"], downcast="float")

display(sentiment_data)
```

```python
# Creates a combined violinplot and stripplot to show the sentiment 
# polarity of tweets where data is available, divided by day.  
# Note: because of the available data, this plot only displays dates 
# in January where the tweets were able to be analyzed for sentiment
# (e.g. tweets that exist or are still available), and the x-axis
# should be viewed as discrete units represented by dates rather
# than a linear or incremental progression of time.

sns.set_theme(style="whitegrid")
plt.figure(figsize=(30,16))
graph_2= sns.violinplot(x="Date", y="Polarity", data=sentiment_data, inner=None, color=".8")
graph_2= sns.stripplot(x="Date", y="Polarity", data=sentiment_data)
plt.xticks(rotation=80, size=20)
plt.yticks(size=20)
plt.xlabel("Date", size=25)
plt.ylabel("Polarity", size=25)
plt.title("Total Sentiment Polarity By Day", size=30)
plt.show()
```

# Visualization of Average Sentiment Polarity Per Day

```python
# Takes the daily average of sentiment polarity provided by
# Alice and turns it in a DataFrame with columns for date and
# average sentiment polarity. Note: Due to the intent for this
# visualization to display change over time, dates without
# available sentiment data have been appended onto the data
# file with a nil Polarity value to reflect their lack of data.

average_sentiment= pd.read_csv('WORK polarity_per_day.csv', delimiter= ",", header=None)
average_sentiment.columns= ["Date", "Polarity"]
average_sentiment["Polarity"]= pd.to_numeric(average_sentiment["Polarity"], downcast="float")

display(average_sentiment)
```

```python
# Creates a scatterplot showing the average sentiment of tweets 
# for each day in January where there were available tweets to 
# analyze for sentiment.  Dates without markers indicate dates
# where there was no data available.

average= sns.relplot(x="Date", y="Polarity", s=50, data=average_sentiment, height=6, aspect=2)
average.set_xticklabels(rotation=70, size=10)
average.set_xlabels("Date", size=15)
average.set_ylabels("Polarity", size=15)
plt.title("Average Sentiment Polarity Per Day", size=20)
plt.show()
```
